# Feature Engineering & Risk Signal Discovery

This notebook examines the transformed staging layer (`stg_fl_school_grades`) to identify early risk signals associated with graduation outcomes.

The objective is to validate engineered features, assess data integrity, and surface candidate predictors for a downstream binary risk classification task.

This analysis serves as the bridge between data transformation and model development.

## 1. Analytical Context

Graduation rates is treated as a downstream outcome variable.

Rather than modeling the exact graduation rate, the analysis focuses on identifying schools that may be at elevated risk of underperformance.

This perspective aligns with real-world decision support use cases, where the goal is early identification and intervention rather than precise point estimation.

```python
import duckdb

con = duckdb.connect("../accountability_project/dev.duckdb")

query = """
select
    district_number,
    district_name,
    school_number,
    school_name,
    graduation_rate_2023_24
from stg_fl_school_grades
where graduation_rate_2023_24 is not null
order by graduation_rate_2023_24 asc
limit 10
"""

con.execute(query).fetchdf()
```

## 2. Graduation Rate Validation

The lowest observed graduation rates are examined to verify that:
- Values are correctly cast and interpretable
- Extreme values reflect real entities rather than data quality issues
- Stuctural patterns begin to emerge across school categories

Several of the lowest-performing schools are virtual or non-traditional, suggesting that school structure may act as a contexual risk modifier.

```python
con.close()
```

## 3. Hypothesis Formation

Initial exploration motivates sever working hypotheses:

- School type may moderate graduation outcomes
- Virtual schools may exhibit greater variance and downside risk
- Socioeconomic disadvantage may correlate with lower graduation rates

These hypotheses guide subsequent feature-level exploration and modeling decisions.

## 4. Feature-level Signal Exploration

Potential predictors are evaluated for their relationship to graduation outcomes, including:

- School type
- Title I status
- Percent of economically disadvantaged students

The analysis emphasizes distributional differences and subgroup behavior, rather than assuming linear or uniform effects across schools.

## 5. Modeling Implications

Observed patterns indicate that graduation risk is unevenly distributed and influenced by structural and socioeconomic context.

These findings motivate a binary risk classification approach with interpretable drivers, rather than a single regression model.

The resulting feature set is designed to support explainibility and downstream decision-making.

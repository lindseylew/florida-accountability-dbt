# Florida School Accountability Data - Raw Data Preparation

This notebook prepares the 2025 Florida School Grades dataset for ingestion into as structured analytics pipeline (dbt + DuckDB).

Objective:
- Extract structured tabular data from a semi-formatted Excel source
- Standardize headers
- Remove formatting artifacts
- Export a clean CSV for reproducible warehouse ingestion

## Data Source

Source: Florida Department of Education - 2025 School Grades

The raw Excel file contains:
- Multi-row headers
- Formatting rows
- Blank columns
- Mixed data types
- Embedded metadata rows

This notebook programmatically extracts the true header row and constructs a clean rectangular dataset.

## Load Raw Excel File

The dataset does not begin on row 0.
The rows will be formatted and the true header row will be located dynamically.

```python
import pandas as pd

path = "/Users/lindseylewis/Downloads/SchoolGrades25.xlsx"

raw = pd.read_excel(path, header=None, skiprows=2)
```

## Identify and Apply True Header Row

The actual column headers begin at the row containing "District Number".
The header row will be located programmatically to ensure reproducibility.

```python
header_row_idx = raw.index[raw.iloc[:, 0].astype(str).str.strip().eq("District Number")][0]

raw.columns = raw.iloc[header_row_idx].astype(str).str.strip()
df = raw.iloc[header_row_idx + 1 :].copy()
```

## Remove Structural Artifacts
- Drop fully empty columns
- Remove duplicate header rows embedded in the dataset.

```python
df = df.dropna(axis=1, how="all")

df = df[df["District Number"].astype(str).str.strip() != "District Number"]

print(df.head())
print(df.columns.to_list()[:25])
print(df.shape)
```

## Initial Data Audit

Dataset dimensions:
- Rows: 3,465 schools
- Columns: 54 performance and demographic indicators

The dataset includes:
- Academic achievement metrics
- Graduation rate
- Socioeconomic indicators
- Historical grade data
- School type metadata

## Standardize String Fields

Whitespace inconsistencies can cause downstream join and grouping issues.
All string fields are stripped prior to export.


```python
for col in df.columns:
    df[col] = df[col].apply(lambda x: x.strip() if isinstance(x, str) else x)

out_path = "/Users/lindseylewis/Desktop/ML/accountability-dbt-pipeline/florida-accountability-dbt/accountability_project/seeds/fl_school_grades_2025.csv"
```

## Export Clean CSV for Warehouse Ingestion

The cleaned dataset is exported as a dbt seed file.

This enables:
- Version-controlled data ingestion
- Reproducible loading into DuckDB
- Structured transformation via dbt models.

```python
df.to_csv(out_path, index=False)
print("Wrote:", out_path)

df.shape
df.head()
```
